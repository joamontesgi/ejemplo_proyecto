```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv',
                 index_col='Month')
```

```python
df
```

```python
np.random.seed(0)
faltantes = np.random.choice(df.index, size=10, replace=False)
df.loc[faltantes] = np.nan
```

```python
df
```

```python
plt.figure(figsize=(14, 7))
df['Passengers'].plot(title='Serie de tiempo')
plt.xlabel('Fecha')
plt.ylabel('Número de pasajeros')
plt.show()
```

```python
#Imputación se va a realiar con el valor anterior
df_fill = df.fillna(method='ffill')
```

```python
plt.figure(figsize=(14, 7))
df_fill['Passengers'].plot(title='Serie de tiempo')
plt.xlabel('Fecha')
plt.ylabel('Número de pasajeros')
plt.show()
```

```python
#Imputación se va a realiar con el valor anterior
df_bfill = df.fillna(method='bfill')
```

```python
plt.figure(figsize=(14, 7))
df_bfill['Passengers'].plot(title='Serie de tiempo')
plt.xlabel('Fecha')
plt.ylabel('Número de pasajeros')
plt.show()
```

```python
mean = df['Passengers'].mean()
df_mean =df.fillna(mean)
```

```python
plt.figure(figsize=(14, 7))
df_mean['Passengers'].plot(title='Serie de tiempo')
plt.xlabel('Fecha')
plt.ylabel('Número de pasajeros')
plt.show()
```
