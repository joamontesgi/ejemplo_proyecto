```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
import joblib
```

```python
df = pd.read_csv('/content/weather_classification_data.csv')
df
```

```python
X = df.drop('Temperature', axis=1)
y = df['Temperature']
```

```python
X
```

```python
columnas_categoricas = ['Cloud Cover', 'Season', 'Location', 'Weather Type']
```

```python
X_modificado = pd.get_dummies(X, columns=columnas_categoricas)
```

```python
X_modificado
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_modificado, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(mse)
```

```python
# Guardar las columnas
joblib.dump(X_modificado.columns, 'model_columns.pkl')

# Guardar el modelo
joblib.dump(modelo, 'regresion_lineal.pkl')
```
